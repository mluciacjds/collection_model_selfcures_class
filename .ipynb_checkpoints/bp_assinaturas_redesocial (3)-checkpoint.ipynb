# 4. Funções Globais

```python
def plot_data (df, plot_type = 'Bar', plot_title = None, cols_plot = None, cols_labels = None, other_params =None):
    '''
    Input: 
      - Data de Nascimento formato String 'YYYY-mm-dd'
      - Data de Referencia formato String 'YYYY-mm-dd'
    Função: Calcula a idade a partir da diferença entre a data de referência e a data de nascimento
    OutPut: Idade em anos (Integer)
    '''
    
    if (cols_plot is None):
            print ('Erro: Colunas para plot não fornecidas')
            return
        
    if (plot_type == 'Bar'):
        sns.set(rc={'figure.figsize':(8,4)})
        sns.barplot(
                    x=cols_plot[0], 
                    y=cols_plot[1], 
                    color='blue', 
                    data= df[cols_plot]
                    )
        
        plt.xticks(rotation=60)
        if (cols_labels is None):
            plt.xlabel(cols_plot[0])
            plt.ylabel(cols_plot[1])
        else:
            plt.xlabel(cols_labels[0])
            plt.ylabel(cols_labels[1])
            
        if (plot_title is None): 
            plt.title(cols_labels[1])
        else:
            plt.title(plot_title)
        plt.gca().yaxis.grid(linestyle=':')
        return
    
    if (plot_type == 'BoxPlot'):
        if (len (cols_plot)<2):
            sns.set(rc={'figure.figsize':(8,8)})
            sns.boxplot(
                        y=cols_plot[0], 
                        color='blue', 
                        data= df
                        )
        else:
            sns.set(rc={'figure.figsize':(15,15)})
            sns.boxplot( y=df[cols_plot[0]],
                        x=df[cols_plot[1]] 
                        )
        if (plot_title is None): 
            plt.title(cols_labels[0])
        else:
            plt.title(plot_title)
        return
    
    if (plot_type == "DistPlot"):
        if (other_params is None):
            nbins = 4
        else:
            nbins = other_params[0]
        sns.set(rc={'figure.figsize':(10,10)})
        sns.displot(
                   df, 
                   x=cols_plot[0], 
                   bins=nbins)
        plt.xticks(rotation=60)
        if (plot_title is None): 
            plt.title(cols_labels[0])
        else:
            plt.title(plot_title)
        return
    return
```

```python
def calcular_datediff(col_data_inicial, dreferencia = "2020-05-31", unidade = 'years'):
    '''
    Input: 
      - Data Inicial formato String 'YYYY-mm-dd'
      - Data de Referencia formato String 'YYYY-mm-dd'
      - unidade: ['anos', 'meses', 'dias']
    Função: Calcula a diferença entre duas datas
    OutPut: Diferença em anos, meses ou dias (Integer ou Float)
    '''
    if (unidade == 'years'):
        return relativedelta(datetime.strptime(dreferencia, "%Y-%M-%d"), datetime.strptime(col_data_inicial, "%Y-%M-%d")).years
    elif (unidade == 'months'):
        return relativedelta(datetime.strptime(dreferencia, "%Y-%M-%d"), datetime.strptime(col_data_inicial, "%Y-%M-%d")).months
    else:
        days_diff = (datetime.strptime(dreferencia, "%Y-%m-%d") - datetime.strptime(col_data_inicial, "%Y-%m-%d")).days
        if (days_diff <0 ):
            return 0
        else:
            return days_diff
        #return relativedelta(datetime.strptime(dreferencia, "%Y-%M-%d"), datetime.strptime(col_data_inicial, "%Y-%M-%d")).days
```

```python
def get_days_fromdiff(col_):
    try:
        new_val = str(col_)
        new_val = int(new_val.split(' ')[0])
        return new_val
    except:
        return None
```

```python
def set_account_status ( col_days_til_ref):
    if (col_days_til_ref >=15):
        return "Inativa"
    else:
        return "Ativa"
```

    - Parametros de Class Imbalance
   

```python
def score_model(model, params, X_train, y_train, oversampler = 'smote', cv=None):
    """
    Creates folds manually, and upsamples within each fold.
    Returns an array of validation (recall) scores
    """
    if (cv is None):
        cv = KFold(n_splits=5,shuffle = True, random_state=1)
    if (oversampler == 'smote'):
        smoter = SMOTE(sampling_strategy='auto',  
                        random_state=0,  
                        k_neighbors=5,
                        n_jobs=4)
    else: 
        params['class_weight'] = 'balanced'
        
    best_score = -100
    best_model = ''
    for train_fold_index, val_fold_index in cv.split(X_train, y_train):
       
        X_train_fold, y_train_fold = X_train_feat_sel.filter(items = train_fold_index, axis=0), y_train.filter(items = train_fold_index, axis=0)
        X_val_fold, y_val_fold = X_train_feat_sel.filter(items =val_fold_index, axis=0), y_train.filter(items =val_fold_index, axis=0)

        # Upsample only the data in the training section
        if (oversampler == 'smoter'):
            X_train_fold, y_train_fold = smoter.fit_resample(X_train_fold,y_train_fold)
    
        # Fit the model on the upsampled training data
        model_obj = model(**params).fit(X_train_fold, y_train_fold)
        # Score the model on the (non-upsampled) validation data
        score = recall_score(y_val_fold, model_obj.predict(X_val_fold))
        if (score > best_score):
            best_score = score
            best_model = model_obj
    
    return  best_model, best_score

# Example of the model in action
```

```python
parametros = {
                'C':0.5,
                'penalty': 'l2',
                'max_iter': 300, 
                'random_state': 10
                }
bestmodel, bestscore = score_model(LogisticRegression, parametros, X_train_feat_sel, y_train)
```
