```
def esta_no_periodo(string_lista, periodo_inicial_d, periodo_final_d):
  string_match_just_date = re.match(r'.*maua.*_exec_(.*)_.*_.*_.*.xls', string_lista)
  try:
    data_nome_lista = datetime.strptime(string_match_just_date.group(1), "%Y-%m-%d")
  except:
    return 0
  else:
    if ((data_nome_lista >= periodo_inicial_d) and (data_nome_lista <= periodo_final_d)):
      return 1
    return 0

def get_listas_periodo_mes_execucao(periodo_inicial, periodo_final):
  periodo_inicial_d = datetime.strptime(periodo_inicial, "%Y-%m-%d")
  periodo_final_d = datetime.strptime(periodo_final, "%Y-%m-%d")
  l = os.listdir("/dbfs/mnt/datalake-ambiental/output/modelo_self_cure/maua/listas")
  l = list(filter(lambda x: '.xls' in x, l))
  l = list(map(lambda x:"dbfs:/mnt/datalake-ambiental/output/modelo_self_cure/maua/listas/"  + x, l))
  l = list(filter(lambda x: esta_no_periodo(x, periodo_inicial_d, periodo_final_d), l))
  return l

"""
def get_listas_periodo_mes_execucao(periodo_inicial, periodo_final):
  periodo_inicial_d = datetime.strptime(periodo_inicial, "%Y-%m-%d")
  periodo_final_d = datetime.strptime(periodo_final, "%Y-%m-%d")
  l = list(dbutils.fs.ls("dbfs:/mnt/datalake-ambiental/output/modelo_self_cure/maua/listas"))
  l = list(map(lambda x:x.path, l))
  l = list(filter(lambda x: '.xls' in x, l))
  l = list(filter(lambda x: esta_no_periodo(x, periodo_inicial_d, periodo_final_d), l))
  return l
"""
```

```
replace_str_f =udf(lambda x: x.replace(",","."), StringType())
```

```
def get_dataframe_unico_listas2(listas_):
  
  dict_listas = dict()
  dict_listas["paracortar"]={}
  dict_listas["parafiscalizar"] = {}
  
  for element in listas_:
    string_match_complete_datetime = re.match(r'.*_exec_(.*).xls', element)
    string_match_just_date = re.match(r'.*_exec_(.*)_.*_.*_.*.xls', element)
    
    if (string_match_just_date):
      range_inferior = datetime.strptime(string_match_just_date.group(1) + "_10:00:00","%Y-%m-%d_%H:%M:%S")
      range_superior = datetime.strptime(string_match_just_date.group(1) + "_11:50:00", "%Y-%m-%d_%H:%M:%S")
      
      if (string_match_complete_datetime):
        actual_date = datetime.strptime(string_match_complete_datetime.group(1), "%Y-%m-%d_%H_%M_%S")
       
        if "paracortar" in element:
          if (not(string_match_just_date.group(1) in dict_listas["paracortar"])):
            if ((actual_date >= range_inferior) and (actual_date <= range_superior)):
              dict_listas["paracortar"][string_match_just_date.group(1)] = element
          else:
            string_compare = re.match(r'.*\_exec_(.*).xls', dict_listas["paracortar"][string_match_just_date.group(1)])
            date_compare = datetime.strptime(string_compare.group(1), "%Y-%m-%d_%H_%M_%S")
            if ((actual_date >= range_inferior) and (actual_date <= range_superior) and (actual_date>date_compare)):
              dict_listas["paracortar"][string_match_just_date.group(1)] = element
        elif "parafiscalizar" in element:
          if (not(string_match_just_date.group(1) in dict_listas["parafiscalizar"])):
            if ((actual_date >= range_inferior) and (actual_date <= range_superior)):
              dict_listas["parafiscalizar"][string_match_just_date.group(1)] = element
          else:
            string_compare = re.match(r'.*\_exec_(.*).xls', dict_listas["parafiscalizar"][string_match_just_date.group(1)])
            date_compare = datetime.strptime(string_compare.group(1), "%Y-%m-%d_%H_%M_%S")
            if ((actual_date >= range_inferior) and (actual_date <= range_superior) and (actual_date>date_compare)):
              dict_listas["parafiscalizar"][string_match_just_date.group(1)] = element
       
  columns_paracortar = ['ID_PROXIMIDADE', 'GRUPO_LEITURA', 'SETOR_LEITURA', 'CDC', 'CLASSIFICACAO_CDC', 'TOTAL_DEBITO_CDC', 'TOTAL_FATURAS_DEBIDAS_CDC', 'lista', 'data', 'row_id', 'DEBITO_ROW']

  vals = [('NONE', 'NONE','NONE', 1, 'x', 50.0, 50,'qq', datetime.strptime("2018-01-01", "%Y-%m-%d"), 1, 0.0)]
  df_base = sqlContext.createDataFrame(vals, columns_paracortar)
  
  for k,v in dict_listas["paracortar"].items():
    try:
      x = pd.read_excel(v.replace('dbfs:/','/dbfs/'))
      x.to_csv(v.replace('dbfs:/','/dbfs/').replace('.xls','.csv'))
    except:
      print("nao conseguiu salvar:", v.replace('dbfs:/','/dbfs/').replace('.xls','.csv'))
      pass
    else:
      try:
        df_read = spark.read.format('csv').option('delimiter',',')\
        .option('encoding', 'ISO-8859-1').option('header', 'true')\
        .load(v.replace('dbfs:/','dbfs:/').replace('.xls','.csv'))\
        .withColumn('CDC', col('CDC').cast(IntegerType()))
      except:
          print("nao conseguiu ler:", v.replace('dbfs:/','dbfs:/').replace('.xls','.csv'))
          pass
      else:
        df_for_union= df_read.select('ID_PROXIMIDADE', 'GRUPO_LEITURA', 'SETOR_LEITURA', "CDC", "CLASSIFICACAO_CDC", "TOTAL_DEBITO_CDC", "TOTAL_FATURAS_DEBIDAS_CDC")\
        .withColumn("lista", lit(k)).withColumn("data",lit(k))\
        .withColumn("TOTAL_DEBITO_CDC", replace_str_f(col('TOTAL_DEBITO_CDC')))
        df_for_union = df_for_union.withColumn("row_id", monotonically_increasing_id())
        df_for_union = df_for_union.withColumn('unique_label', lit('A'))
        windowval = (Window.partitionBy('unique_label').orderBy('row_id').rangeBetween(Window.unboundedPreceding, 0))
        df_for_union = df_for_union.withColumn('DEBITO_ROW', round(sum('TOTAL_DEBITO_CDC').over(windowval),2))
        df_for_union = df_for_union.drop('unique_label')
        df_base = df_base.union(df_for_union)
        
  df_base = df_base.where(~(col('lista') == "qq"))
  df_base = df_base.withColumn('data', col('data').cast(DateType()))
  df_base = df_base.withColumn('TOTAL_DEBITO_CDC', col('TOTAL_DEBITO_CDC').cast(FloatType()))
  df_base = df_base.withColumn('DEBITO_ROW', col('DEBITO_ROW').cast(FloatType()))
  df_base = df_base.withColumn('mes_execucao', month(col('data')))
  df_base = df_base.withColumn('ano_execucao', year(col('data')))
  df_base = df_base.withColumn('dia_execucao', dayofmonth(col('data')))
  return  df_base, dict_listas
```

```
def concat_df_listas_modelo(df_modelo_manha, df_modelo_tarde):
  
  df_modelo_1 = df_modelo_manha.withColumn('HIT_CDC', col('TOTAL_DEBITO_CDC')/col('TOTAL_FATURAS_DEBIDAS_CDC'))\
                               .select("ID_PROXIMIDADE", "GRUPO_LEITURA", "SETOR_LEITURA",\
                                       "CDC", "TOTAL_DEBITO_CDC", "HIT_CDC", "row_id", "data",\
                                       "dia_execucao", "mes_execucao", "ano_execucao")
  
  df_modelo_2 = df_modelo_tarde.select("ID_PROXIMIDADE", "ID_GRUPO_LEITURA", "ID_SETOR_LEITURA",\
                                       "CDC", "TOTAL_DEBITO_CDC", "HIT_CDC", "row_id", "data",\
                                       "dia_execucao", "mes_execucao", "ano_execucao")\
                               .withColumnRenamed("ID_GRUPO_LEITURA", "GRUPO_LEITURA")\
                               .withColumnRenamed("ID_SETOR_LEITURA", "SETOR_LEITURA")
  
  result = df_modelo_1.union(df_modelo_2)
  return result
```

```
def uso_mensal_modelo(periodo_inicial, periodo_final):
  
  df_listas_paracortar = spark.read.table("modelos_preditivos.kpis_maua_maua_listas_modelo")
  df_cortes = spark.read.table('modelos_preditivos.kpi_maua_maua_cortes_religacoes_')
  
  df_cortes = df_cortes.withColumn('ano_corte', year(col('DATA_EXECUCAO_CORTE')))\
                       .withColumn('mes_corte', month(col('DATA_EXECUCAO_CORTE')))\
                       .withColumn('dia_corte', dayofmonth(col('DATA_EXECUCAO_CORTE')))
  
  df_cortes = df_cortes.where((col('DATA_EXECUCAO_CORTE') >= datetime.strptime(periodo_inicial, "%Y-%m-%d"))\
                              & (col('DATA_EXECUCAO_CORTE') <= datetime.strptime(periodo_final, "%Y-%m-%d")))
  
  df_cortes = df_cortes.where((col('DATA_EXECUCAO_CORTE').isNotNull()))
  
  resumo_mensal_listas_paracortar = df_listas_paracortar.groupBy('ano_execucao', 'mes_execucao', 'CDC').agg({'total': 'count'})\
                                                        .withColumnRenamed('CDC', 'CDC_lista')
  
  resumo_mensal_cortes = df_cortes.groupBy('ano_corte', 'mes_corte', 'CDC').count()
  
  resumo_mensal = resumo_mensal_cortes.alias('a').join(resumo_mensal_listas_paracortar.alias('b'),\
                                                       [resumo_mensal_cortes.CDC == resumo_mensal_listas_paracortar.CDC_lista,\
                                                        resumo_mensal_cortes.ano_corte == resumo_mensal_listas_paracortar.ano_execucao,\
                                                        resumo_mensal_cortes.mes_corte == resumo_mensal_listas_paracortar.mes_execucao],\
                                                       "LEFT")\
                                                  .select('a.*', 'b.CDC_lista')
  
  cortados_mensal = resumo_mensal.groupBy('ano_corte', 'mes_corte').agg({'total_cortados': 'count'})
  
  cortados_lista_mensal = resumo_mensal.where(col('CDC_lista').isNotNull()).groupBy('ano_corte', 'mes_corte').agg({'total_cortados_modelo': 'count'})
  
  cortados = cortados_mensal.alias('m').join(cortados_lista_mensal.alias('l'), \
                                             [cortados_mensal.ano_corte == cortados_lista_mensal.ano_corte, \
                                             cortados_mensal.mes_corte == cortados_lista_mensal.mes_corte],\
                                             'LEFT')\
                                        .select('m.*', 'l.total_cortados_modelo')
  
  return cortados
  
```

```
def uso_diario_modelo(periodo_inicial, periodo_final):
  
  df_listas_paracortar = spark.read.table('modelos_preditivos.kpis_maua_maua_listas_modelo_campo')
  df_cortes = spark.read.table('modelos_preditivos.kpi_maua_maua_cortes_religacoes_')
  
  df_cortes = df_cortes.withColumn('ano_corte', year(col('DATA_EXECUCAO_CORTE'))).withColumn('mes_corte', month(col('DATA_EXECUCAO_CORTE'))).withColumn('dia_corte', dayofmonth(col('DATA_EXECUCAO_CORTE')))
  
  df_cortes = df_cortes.where((col('DATA_EXECUCAO_CORTE') >= datetime.strptime(periodo_inicial, "%Y-%m-%d")) & (col('DATA_EXECUCAO_CORTE') <= datetime.strptime(periodo_final, "%Y-%m-%d")))
  
  df_cortes = df_cortes.where((col('DATA_EXECUCAO_CORTE').isNotNull()))
  
  df_cortes = df_cortes.alias('pc').join(df_listas_paracortar.alias('pcortar'), [df_cortes.idligacao == df_listas_paracortar.CDC, df_cortes.ano_corte == df_listas_paracortar.ano_execucao, df_cortes.mes_corte == df_listas_paracortar.mes_execucao, df_cortes.dia_corte == df_listas_paracortar.dia_execucao], "LEFT").select('pc.*', 'pcortar.*')
  
  df_cortes.cache()

  window_spec_ano_mes_dia = Window.partitionBy('ano_corte', 'mes_corte', 'dia_corte') 
  
  df_cortes = df_cortes.withColumn('cortado_modelo', when(col('lista').isNotNull(), col('idligacao')).otherwise(None))
  
  df_cortes = df_cortes.withColumn('cortados', size(collect_set(col('idligacao')).over(window_spec_ano_mes_dia)))
  
  df_cortes = df_cortes.withColumn('cortados_modelo', size(collect_set(col('cortado_modelo')).over(window_spec_ano_mes_dia)))

  #df_cortes = df_cortes.withColumn('cortados_modelo', sum(col('cortado_modelo')).over(window_spec_ano_mes_dia))
  
  df_cortes = df_cortes.withColumn('cortados_prioridade', sum(when((col('lista').isNotNull()) & (col('row_id')< col('cortados')),1).otherwise(0)).over(window_spec_ano_mes_dia))
  
  indicadores_cortes = df_cortes.select('ano_corte', 'mes_corte', 'dia_corte', 'cortados', 'cortados_modelo', 'cortados_prioridade').dropDuplicates(['ano_corte', 'mes_corte', 'dia_corte'])

  return indicadores_cortes, df_cortes
```

```
def uso_diario_modelo_debitosugestao(periodo_inicial, periodo_final):
  
  df_listas_paracortar = spark.read.table('modelos_preditivos.kpis_maua_maua_listas_modelo')
  processos_cortes = spark.read.table('modelos_preditivos.kpi_maua_maua_processos_')
  processos_cortes = processos_cortes\
  .withColumn('ano_corte', year(col('data_corte')))\
  .withColumn('mes_corte', month(col('data_corte')))\
  .withColumn('dia_corte', dayofmonth(col('data_corte')))
  
  processos_cortes = processos_cortes.where((col('data_corte') >= datetime.strptime(periodo_inicial, "%Y-%m-%d")) & (col('data_corte') <= datetime.strptime(periodo_final, "%Y-%m-%d")))
  
  debito_cortados = processos_cortes.groupBy('ano_corte', 'mes_corte', 'dia_corte')\
  .agg(sum(col('valor_fat_proc')).alias('debito_cortado'))\
  .orderBy('ano_corte', 'mes_corte', 'dia_corte')

  processos_cortes = processos_cortes.alias('procs_cortes')\
  .join(df_listas_paracortar.alias('modelo'), \
        [processos_cortes.cdc_proc_lig == df_listas_paracortar.CDC, \
         processos_cortes.ano_corte == df_listas_paracortar.ano_execucao, \
         processos_cortes.mes_corte == df_listas_paracortar.mes_execucao, \
         processos_cortes.dia_corte == df_listas_paracortar.dia_execucao], "LEFT")\
  .select('procs_cortes.*', 'modelo.lista', 'modelo.DEBITO_ROW', 'modelo.TOTAL_DEBITO_CDC',\
          'modelo.row_id', 'modelo.ano_execucao', 'modelo.mes_execucao', 'modelo.dia_execucao')
  
  window_spec = Window.partitionBy('ano_corte', 'mes_corte', 'dia_corte')
  
  processos_cortes = processos_cortes.withColumn('cortados', size(collect_set(col('cdc_proc_lig')).over(window_spec)))
  
  cortados_mes = processos_cortes.select('ano_corte', 'mes_corte', 'dia_corte', 'cortados')\
  .dropDuplicates(['ano_corte', 'mes_corte', 'dia_corte'])
  
  debito_sugestao = df_listas_paracortar.alias('listas')\
    .join(cortados_mes.alias('cortesmes'), \
          [df_listas_paracortar.ano_execucao == cortados_mes.ano_corte,\
           df_listas_paracortar.mes_execucao == cortados_mes.mes_corte,\
           df_listas_paracortar.dia_execucao == cortados_mes.dia_corte, 
           df_listas_paracortar.row_id == cortados_mes.cortados],'LEFT')\
    .select('listas.ano_execucao', 'listas.mes_execucao', 'listas.dia_execucao','listas.row_id', 'listas.DEBITO_ROW', 'listas.TOTAL_DEBITO_CDC', 'cortesmes.cortados')
    
  window_spec_listas = Window.partitionBy('ano_execucao', 'mes_execucao', 'dia_execucao')
  
  debito_sugestao = debito_sugestao.withColumn('debito_sugestao', max(col('DEBITO_ROW')).over(window_spec_listas))
  
  debito_sugestao = debito_sugestao.dropDuplicates(['ano_execucao', 'mes_execucao', 'dia_execucao'])
  
  debito_sugestao = debito_sugestao.select('ano_execucao', 'mes_execucao', 'dia_execucao', 'debito_sugestao')
  
  cortado_sugerido = debito_cortados.alias('cortados').join(debito_sugestao.alias('sugestao'),\
                                                            [debito_cortados.ano_corte == debito_sugestao.ano_execucao,\
                                                             debito_cortados.mes_corte == debito_sugestao.mes_execucao,\
                                                             debito_cortados.dia_corte == debito_sugestao.dia_execucao], "LEFT")\
  .select('cortados.ano_corte', 'cortados.mes_corte', 'cortados.dia_corte', 'cortados.debito_cortado', 'sugestao.debito_sugestao')\
  .orderBy('cortados.ano_corte', 'cortados.mes_corte', 'cortados.dia_corte')
  
  cortado_sugerido = cortado_sugerido.withColumn('debito_sugestao', when(col('debito_sugestao').isNull(), lit(0)).otherwise(col('debito_sugestao')))

  return cortado_sugerido
  
```

```
df_cortes2 = uso_diario_modelo_debitosugestao(global_periodo_inicial, global_periodo_final)
```
